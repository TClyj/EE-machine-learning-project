```python
# EE399 HW 1
# Yijia Lu
# Student id: 1961366
# 4/9 2023

import numpy as np
import scipy.optimize as opt
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
# (i)

X=np.arange(0,31)
Y=np.array([30, 35, 33, 32, 34, 37, 39, 38, 36, 36, 37, 39, 42, 45, 45, 41,
40, 39, 42, 44, 47, 49, 50, 49, 46, 48, 50, 53, 55, 54, 53])

def f(x, A, B, C, D):
    return A * np.cos(B*x) + C*x + D

# Fit the model to the data
popt, pcov = curve_fit(f, X, Y)
task1_graph = f(X, *popt)

A = popt[0]
B = popt[1]
C = popt[2]
D = popt[3]
print('A =', A)
print('B =', B)
print('C =', C)
print('D =', D)
print('Minimum error =', np.sqrt(np.sum((f(X, *popt) - Y)**2) / len(X)))

plt.plot(X, Y, 'o', label='Data points')
plt.plot(X, task1_graph, 'r', label=f'Data predicted by the model')
plt.legend()
plt.show()
```

```python
# (ii) 1/6

# Define the two fixed parameters
A_fix = A
B_fix = B

# Define the range of values for the two swept parameters
C_range = np.linspace(C - 10, C + 10, 100)
D_range = np.linspace(D - 10, D + 10, 100)

# Create a grid of parameter values
C_grid, D_grid = np.meshgrid(C_range, D_range)

# Compute the error for each combination of parameter values
error_grid = np.zeros((len(C_range), len(D_range)))
for i in range(len(C_range)):
    for j in range(len(D_range)):
        error_grid[i,j] = np.sqrt(np.mean((f(X, A_fix, B_fix, C_range[i], D_range[j]) - Y)**2))

# Visualize the results in a 2D color plot
plt.figure()
plt.pcolor(C_grid, D_grid, error_grid)
plt.colorbar()
plt.xlabel('C')
plt.ylabel('D')
plt.title('Error Landscape with fixed A and B')
plt.show()
```

```python
# (iii)

# Split the data into training and test sets
X_train = X[:20]
Y_train = Y[:20]
X_test = X[20:]
Y_test = Y[20:]

# Fit a line to the training data
coeffs_line = np.polyfit(X_train, Y_train, 1)
Y_line_train = np.polyval(coeffs_line, X_train)
Y_line_test = np.polyval(coeffs_line, X_test)
error_line_train = np.sqrt(np.sum((Y_line_train - Y_train)**2) / len(X_train))
error_line_test = np.sqrt(np.sum((Y_line_test - Y_test)**2) / len(X_test))

# Fit a parabola to the training data
coeffs_parabola = np.polyfit(X_train, Y_train, 2)
Y_parabola_train = np.polyval(coeffs_parabola, X_train)
Y_parabola_test = np.polyval(coeffs_parabola, X_test)
error_parabola_train = np.sqrt(np.sum((Y_parabola_train - Y_train)**2) / len(X_train))
error_parabola_test = np.sqrt(np.sum((Y_parabola_test - Y_test)**2) / len(X_test))

# Fit a 19th degree polynomial to the training data
coeffs_poly = np.polyfit(X_train, Y_train, 19)
Y_poly_train = np.polyval(coeffs_poly, X_train)
Y_poly_test = np.polyval(coeffs_poly, X_test)
error_poly_train = np.sqrt(np.sum((Y_poly_train - Y_train)**2) / len(X_train))
error_poly_test = np.sqrt(np.sum((Y_poly_test - Y_test)**2) / len(X_test))

# Print the training errors
print('Training errors')
print('Line: {:.2f}'.format(error_line_train)) # use {:.2f} to keep two decimal place
print('Parabola: {:.2f}'.format(error_parabola_train))
print('19th degree polynomial: {:.2f}'.format(error_poly_train))

# Print the test errors
print('Test errors')
print('Line: {:.2f}'.format(error_line_test))
print('Parabola: {:.2f}'.format(error_parabola_test))
print('19th degree polynomial: {:.2f}'.format(error_poly_test))

# Plot the results
plt.scatter(X_train, Y_train, label='Training data')
plt.scatter(X_test, Y_test, label='Test data')
plt.plot(X_train, Y_line_train, label='Line')
plt.plot(X_train, Y_parabola_train, label='Parabola')
plt.plot(X_train, Y_poly_train, label='19th degree polynomial')
plt.legend()
plt.show()
```
